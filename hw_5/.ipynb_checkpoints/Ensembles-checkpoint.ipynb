```python
# Сгенерируем какую-нибудь необычную зависимость и научимся ее предсказывать
np.random.seed(42)
m = 500
X = np.linspace(-3, 3, m).reshape(-1, 1)
y = (3 + 2/np.pi * np.arcsin(np.cos(10 * X))) * X
y = y + np.random.randn(m, 1) / 3
plt.plot(X.reshape(-1), y.reshape(-1), "b.");
```

```python
# Функция для отрисовки предсказаний деревьев решений в случае регрессии
def plot_regression_predictions(tree_reg, X, y, label, sub, axes=[-3, 3, -10, 10]):
    # extract test data and vars: implementated the external 
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    y1 = (3 + 2/np.pi * np.arcsin(np.cos(10 * x1))) * x1
    y1 = y + np.random.randn(500, 1) / 3
    
    y_pred = tree_reg.predict(x1)
    
    #if ylabel:
    #    plt.ylabel(ylabel, fontsize=18, rotation=0)
    
    # plot on facet according to axes
    sub.set_xlabel("$x_1$", fontsize=18)
    sub.set_ylabel(r"$\hat{y} $", fontsize=18, rotation=0)
    sub.set_title(f'Decision tree regression model with depth = {label}', fontsize=12)

    sub.plot(x1.reshape(-1), y1.reshape(-1), "blue")
    sub.plot(x1, y_pred, linewidth=1.5, color='red')
```

Изобразите, как меняются предсказания дерева при увеличении максимальной глубины (можно взять что-то в диапазоне от 1 до 15):

```python
# Assign depth which will be analyzed:
depth = range(1,15)
# Make templat for facet plot
fig, axes = plt.subplots(5, 3, figsize=(25, 25))
axes = axes.flatten()

# Plot each depth-figure on axes
for i, d in enumerate(depth):
    tr = DecisionTreeRegressor(criterion = 'mse', max_depth = d)
    tr.fit(X,y)
    plot_regression_predictions(tr, X, y, label=d, sub=axes[i])
    
```

We see that with the depth = 1 our regression model predicts the form of distribution far from real. While rising the depth we come over with the sense that on 3th depth step our division is more likely the real one. Even it i not sufficient to outliers and varians, the overal form is predicted properly. 

With the following depth increase **on the 12th step** exactly **the variance** starts to be predicted <u>the best</u> and then practically no improvements can be noticed in variance distribution. However <u>here</u> our model is more taking attraction to the **dispersion** than to the average distribution, meaning that it would <u>tend to predict</u> estimations for peculiar variables **differing from the majority**. And **the majority** would be described <u>worse</u>.

Thus, it seems that **the optimal depth** for DT regression model (on this data) is **3**, as from the 4th step of depth our model more highlights outliers.

Представим, что у нас есть 101 классификатор. Каждый может с вероятностью `p` (равной для всех моделей) правильно предсказать класс объекта. Будем делать предсказания по большинству голосов (majority vote). Постройте зависимость вероятности правильно классифицировать объект от значения `p`. Вам может быть полезная следующая формула:

$$ \large \mu = \sum_ {i = 51} ^ {101} C_{101} ^ ip ^ i (1-p) ^ {101-i} $$

```python
## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*##
import numpy as np
import math as mt
```

```python
X = np.arange(0.0, 1.0, 0.0001).tolist()
y = []
for i in X:
    mu = 0
    for j in range(51,101+1):
        mu += (mt.factorial(101)/(mt.factorial(j)*mt.factorial(101-j))) * (i**j) * ((1.0 - i)**(101-j))
    y.append(mu)
```

```python
plt.xlabel('Individual probability', fontsize = 15)
plt.ylabel('Majority vote probability of success', fontsize = 15)
plt.plot(X,y)
```

```python
## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*##
```

```python
X = range(2, 101, 2)
y = []
for k in X:
    mu = 0
    for n in range(k//2, k + 1):
        mu += (mt.factorial(k)/(mt.factorial(n)*mt.factorial(k-n))) * (0.65**n) * ((1.0 - 0.65)**(k-n))
    y.append(mu)
plt.xlabel('Number of models', fontsize = 15)
plt.ylabel('Majority vote probability of success', fontsize = 15)
plt.plot(X,y)
```

To sum up, it could be mentioned that:

#### Задание 2. 2. Реализация простого RF

**4 балла**

Реализуйте свой собственный класс `RandomForestClassifierCustom`, используя в качестве базовой модели `DecisionTreeClassifier` из `sklearn`.

Небольшое описание:
- Используйте приведенный ниже код
- В методе `fit` в цикле (`i` от 0 до `n_estimators-1`):
    * Зафиксируйте генератор случайных чисел следующим образом np.random.seed(`random_state + i`). Идея в том, что на каждой итерации у нас будет новое значение для генератора случайных чисел, что добавит побольше "случайности", но в то же время мы сможем иметь воспроизводимые результаты
    * После чего выберите `max_features` признаков **без возвращения/without replacement**, сохраните список выбранных признаков (их индексов) в `self.feat_ids_by_tree`
    * Также создайте псевдовыборку при помощи бутстрэпа (выбор **с возвращением/with replacement**) из тренировочных данных. Может помочь функция `np.random.choice` и ее аргумент `replace`
    * Обучите дерево решений с параметрами, заданными в конструкторе класса `max_depth`, `max_features` и `random_state` на полученной псевдовыборке.
- Метод `fit` должен возвращать текущий экземпляр класса `RandomForestClassifierCustom`, то есть `self` (все по-взрослому, как в `sklearn`)
- В методе `predict_proba` мы должны пройти циклом по всем деревьям. Для каждого предсказания, нам нужно будет брать только те признаки, на которых училось изначальное дерево, поэтому мы и сохраняли эту информацию в артрибуте `self.feat_ids_by_tree`. Этот метод должен возвращать предсказанные вероятности (можно делать двумя способами: для каждого дерева предсказывать значение при помощи метода `predict_proba` и потом усреднять эти вероятности, или к примеру пользоваться методом `predict` и также считать среднее.

```python
class RandomForestClassifierCustom(BaseEstimator):
    def __init__(
        self, n_estimators=10, max_depth=None, max_features=None, random_state=SEED
    ):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.max_features = max_features
        self.random_state = random_state
        self.trees = []
        self.feat_ids_by_tree = []

    def fit(self, X, y):
        self.classes_ = sorted(np.unique(y))
        for i in range(0, self.n_estimators - 1):
            # make each tree randomly chosen
            np.random.seed(self.random_state + i)
            # get feature indices
            feat_idx = np.random.permutation(X.shape[1])[:self.max_features]
            # assign which feature is now taken into account
            self.feat_ids_by_tree.append(feat_idx)
            # boot_idx for boosting
            boot_idx = np.random.choice(X[:,feat_idx].shape[0], X[:,feat_idx].shape[0], replace = True)
            # get boosting sets
            X_boot, y_boot = X[:,feat_idx][boot_idx], y[boot_idx]
            # create and train DT model
            tree = DecisionTreeClassifier(max_depth = self.max_depth, max_features = self.max_features, random_state = self.random_state)
            tree.fit(X_boot, y_boot)
            self.trees.append(tree)
        ## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*##
        return self

    def predict_proba(self, X):
        predicts = []
        ## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*##
        # iterate over trees
        for i in range(len(self.trees)):
            f_idx = self.feat_ids_by_tree[i]
            predicts.append(self.trees[i].predict_proba(X[:,f_idx]))
        return np.array(sum(predicts)/len(predicts))
    
    def predict(self, X):
        probas = self.predict_proba(X)
        predictions = np.argmax(probas, axis=1)
        
        return predictions
```

Протестируем нашу реализацию на искусственных данных. Визуализируйте разделяющую границу, которую рисует ваша модель при помощи функции `plot_decision_boundary` (см. примеры в лекции).

```python
def plot_decision_boundary(clf, X, y, axes=[-1.5, 2.5, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(["#ffdab9","#9898ff", "#4B0082"])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if contour:
        custom_cmap2 = ListedColormap(["#ffdab9", "#4c4c7f", "#4B0082"])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", alpha=alpha)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", alpha=alpha)
    plt.axis(axes)
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
```

#### Задание 2. 3. Корреляция базовых моделей

Как мы выянили на лекции, для того, чтобы bagging работал хорошо, предсказания наших моделей не должны сильно коррелировать. Для этого в случайном лесе применяются различные подходы, в том числе и RSM. Давайте посмотрим, как влияет параметр `max_features` на корреляцию базовых моделей в случайном лесу из `sklearn`. В качестве примера будем использовать датасет `breast_cancer`. Для расчета корреляций используйте приведенную ниже функцию `base_model_pair_correlation`. Для каждой модели у вас будет получаться набор значений (попарные корреляции всех деревьев), дальше можно изобразить их в виде боксплотов, как мы на лекции рисовали распределение метрик.

```python
# Функция для расчета попарных корреляций базовых моделей в случайном лесу
def base_model_pair_correlation(ensemble, X):
    corrs = []
    for (i, est1), (j, est2) in combinations(enumerate(ensemble.estimators_), 2):
        Xi_test = X
        Xj_test = X

        ypred_t1 = est1.predict_proba(Xi_test)[:, 1]
        ypred_t2 = est2.predict_proba(Xj_test)[:, 1]

        corrs.append(pearsonr(ypred_t1, ypred_t2)[0])
    return np.array(corrs)
```

```python
# Загрузим данные
breast_cancer = datasets.load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*##
```

```python
pcor = dict()
for i in [1, 2, 5, 8, 10, 15, 20, 30]:
    rf = RandomForestClassifier(max_features = i)
    rf.fit(X_train, y_train)
    pcor[i] = base_model_pair_correlation(rf, X_test)
```

Вы можете заметить, что ансамбль показывает хорошее, но не лучшее качество предсказания, попробуем его улучшить. Как вы знаете, ансамбли работают лучше, когда модели, входящие в них не скоррелированы друг с другом. Определите корреляцию предсказаний базовых моделей в ансамбле на тестовом наборе данных, и удалите из ансамбля те модели, чьи предсказания будут сильнее коррелировать с остальными. Воспользуйтесь функцией `base_model_pair_correlation_for_voting_clf`. **Спойлер**: далеко не факт, что если вы удалите две модели с корреляцией 0.95, то все станет сильно лучше, здесь все будет немного сложнее. Чтобы добиться максимального качества может понадобиться долгий перебор различных комбинаций моделей. Наилучший скор, который мне удалось достичь, это 0.915, но он получен весьма странной комбинацией алгоритмов, а еще и простым перебором всех вариантов)

```python
def base_model_pair_correlation_for_voting_clf(ensemble, X):
    #corrs = []
    n = len(ensemble.estimators_)
    corrs = np.zeros((n, n))
    base_model_names = [f"{est.__class__.__name__}" for est in ensemble.estimators_]
    base_model_ids = [str(est) for est in ensemble.named_estimators_.keys()]
    for (i, est1), (j, est2) in combinations(enumerate(ensemble.estimators_), 2):
        Xi_test = X
        Xj_test = X

        if not isinstance(est1, SVC):
            ypred_t1 = est1.predict_proba(Xi_test)[:, 1]
        else:
            ypred_t1 = est1.decision_function(Xi_test)


        if not isinstance(est2, SVC):
            ypred_t2 = est2.predict_proba(Xi_test)[:, 1]
        else:
            ypred_t2 = est2.decision_function(Xi_test)
        #not convenient output
        #corrs.append((est1, est2, pearsonr(ypred_t1, ypred_t2)[0]))
        # assign to mdel pair-to-pair relationships gained correlation coeff
        corrs[j, i] = pearsonr(ypred_t1, ypred_t2)[0]
        #corrs[]

    return base_model_ids, corrs
```

```python
## ENTER YOUR CODE HERE (/¯◡ ‿ ◡)/¯☆*
# See correlation by heatmap:

## add mask to hide zeros (from seaborn tutorial)
models, corr = base_model_pair_correlation_for_voting_clf(voting_soft, X_test)
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
# plot heatmap, usign Spectral cmap and our model labels
sns.heatmap(corr, xticklabels=models, yticklabels=models, mask = mask, annot = True, cmap = 'Spectral')
```

#### But how to make automatically?

Try to iterate over models and firstly remove each and check the inprovements of cv score and then (after reaching 1) add one by one till improvement:


```python
base_models = [("DT", dt),  ("ETC", etc), ("KNN", knn), 
               ("SVC_LIN", svc_lin), ("SVC_RBF", svc_rbf), ("CAT", cat), ("LGBM", lgbm), 
               ("LGBM_RF", lgbm_rf), ("XGB", xgb), ("XGB_RF", xgb_rf), ("LR", lr), ("NB", nb)]

# reassign selected models (start from all)
bm_selected = [("DT", dt),  ("ETC", etc), ("KNN", knn), 
               ("SVC_LIN", svc_lin), ("SVC_RBF", svc_rbf), ("CAT", cat), ("LGBM", lgbm), 
               ("LGBM_RF", lgbm_rf), ("XGB", xgb), ("XGB_RF", xgb_rf), ("LR", lr), ("NB", nb)]

# default score with all models
voting_soft = VotingClassifier(estimators=bm_selected, voting='soft')
score = cross_val_score(voting_soft, X_train, y_train, cv=3, scoring="f1").mean()

max_iters = 60 # ограничение по итерациям
improve = True # если лучше не станет - выходим из поиска
step = 0
while improve and step < max_iters:
    improve = False
    
    # remove one by one
    for i in range(len(bm_selected)):
        # take all previous and all folowing models (skipping current)
        new_bm = bm_selected[:i] + bm_selected[i+1:]
        voting_soft = VotingClassifier(estimators=new_bm, voting='soft')
        new_score = cross_val_score(voting_soft, X_train, y_train, cv=3, scoring="f1").mean()
        
        if new_score > score:
            improve = True
            bm_selected = new_bm
            score = new_score
            # exit when found best with removing
            break
            
    # try to add to increase       
    for model in base_models:
        # check if we are not going to add existing model
        if model not in bm_selected:
            new_bm = bm_selected + [model] # not append as new list is necessary
            voting_soft = VotingClassifier(estimators=new_bm, voting='soft')
            new_score = cross_val_score(voting_soft, X_train, y_train, cv=3, scoring="f1").mean()

            if new_score > score:
                improve = True
                bm_selected = new_bm
                score = new_score
                break
    step += 1
    


out = pd.DataFrame({'Step': step,
                   'Iterations': improve,
                   'New score': score,
                   'Selected_models': bm_selected})
out
```

Neither with manual combination, nor automatic algorith the cv score higher than ~ 0.86 was not gained. 

In the manual approach firstly  the most correlating models were removed and also RF as we checkedpreviously that it is ueless under incresed correlations (which is seen here). Then left models were analyzed: if we look at heatmap XGB models show pretty high correlation with almost all models. However, the default XGB is rather better than the XGB_RBF, meaning that probably out ensemble could be improved after removing the latter. The same contradiction situation was with LGBM, even more complex, as in this case we were not able to tell exactly the removement of which will improve our ensmble. Thus both LGBM models were removed.
Finally, also the one with less corellation coefficients (KNN and SVC_RBF) were also deleted and it lead to the score equal to 0.859. Which was the hight in this implementation (even under algorithmic approach). 

Algorithmic method under limitation of 60 iterations allowed to gain the score = 0.858.
